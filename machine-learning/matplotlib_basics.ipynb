Imports

```python
import matplotlib.pyplot as plt
import numpy as np
```

Line plot

```python
#line plot
my_array = np.array([0,3,2,6,4,10])
plt.plot(my_array)
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```

Bar Plot (Histogram)

```python
#Bar plot
my_array = np.array([10,30,20,30,50,20])
plt.hist(my_array)
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```

Scatter plot

```python
#scatter plot
my_array_1 = np.array([10,20,30,40,50,30,40,50])
my_array_2 = np.array([60,70,80,90,30,30,40,50])
plt.scatter(my_array_1, my_array_2)
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```
