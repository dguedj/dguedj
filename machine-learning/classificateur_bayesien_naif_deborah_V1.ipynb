```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
#Generate isotropic Gaussian blobs for clustering.
X,y = make_blobs(
  n_samples=500,            #the total number of points equally divided among clusters
  n_features=2,             #The number of features for each sample
  centers=3,             #The number of centers to generate, or the fixed center locations
  cluster_std=1.5,          #The standard deviation of the clusters.
  center_box=(-10.0, 10.0), #The bounding box for each cluster center when centers are generated at random.
  shuffle=True,             #Shuffle the samples.
  random_state=42,        #Determines random number generation for dataset creation
  return_centers=False)     #return the centers of each cluster.

plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.colorbar()
plt.show();
X_train, X_test, y_train, y_test = train_test_split(
                                    X,y,  #X,y   #dataframes
                                    test_size=0.2,  #prévoir 20% des données pour la phase de test
                                    train_size=None, #the proportion of the dataset to include in the train split
                                    random_state=42, #42 #Controls the shuffling applied to the data before applying the split
                                    )
# Représentation des données d'entraînement
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap='viridis', edgecolors='k', marker='o', label='Entraînement')

# Représentation des données de test
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap='viridis', edgecolors='k', marker='x', label='Test')

plt.legend()
plt.title('Données d\'entraînement et de test')
plt.xlabel('Caractéristique 1')
plt.ylabel('Caractéristique 2')

plt.show()

#instanciation
model_Gaussian = GaussianNB()
#training
model_Gaussian.fit(X_train, y_train)

# Prédiction sur les données de test
y_pred = model_Gaussian.predict(X_test)

# Calcul du taux d'erreur
error_rate = 1 - accuracy_score(y_test, y_pred)

print(f"Taux d'erreur : {error_rate * 100:.2f}%")
```
